```python
# Importing essential libraries

import tkinter as tk
from tkinter import Button
from tkinter import ttk
from csv import DictWriter
import os 
from tkcalendar import DateEntry
from datetime import datetime
```

```python
# Title Name of the application
win = tk.Tk()
win.title('Banking Deposit Prediction')    

#Creating Labels for every input field

# Age Label
age_label = ttk.Label(win,text = "Enter the client's Age : ")
age_label.grid(row=0, column = 0, sticky = tk.W)

#Job Label
job_label = ttk.Label(win, text = "Select the client's Job : ")
job_label.grid(row = 1, column = 0, sticky = tk.W)

# Marital Label
mar_label = ttk.Label(win, text = "Select the client's Marital Status")
mar_label.grid(row = 2, column = 0, sticky = tk.W)

# Education Label
edu_label = ttk.Label(win, text = "Select the client's Education")
edu_label.grid(row = 3, column = 0, sticky = tk.W)

# Default Label
def_label = ttk.Label(win, text = "Has the client credit ever Defaulted?")
def_label.grid(row = 4, column = 0, sticky = tk.W)

# Balance Label
bal_label = ttk.Label(win,text = "Enter the client's Balance : ")
bal_label.grid(row=6, column = 0, sticky = tk.W)

# Housing Label
house_label = ttk.Label(win, text = "Does the client have Housing Loan?")
house_label.grid(row = 7, column = 0, sticky = tk.W)

# Loan Label
loan_label = ttk.Label(win, text = "Does the client have Personal Loan?")
loan_label.grid(row = 9, column = 0, sticky = tk.W)

# Contact Label
contact_label = ttk.Label(win, text = "How does does the client like to be contacted?")
contact_label.grid(row = 11, column = 0, sticky = tk.W)

# # Date Label
date_label = ttk.Label(win, text = "Enter last day when the client was contacted")
date_label.grid(row = 12, column = 0)

# Duration Label
duration_label = ttk.Label(win,text = "What was the duration of last contact? ")
duration_label.grid(row=13, column = 0, sticky = tk.W)

# Campaign Label
campaign_label = ttk.Label(win,text = "What was number of contacts done during this duration? ")
campaign_label.grid(row=14, column = 0, sticky = tk.W)

# pdays Label
pdays_label = ttk.Label(win,text = "What was number of days that passed by after the client was last contacted from a previous campaign? ")
pdays_label.grid(row=15, column = 0, sticky = tk.W)

# previous Label
previous_label = ttk.Label(win,text = "What was number of contacts performed before this campaign and for this client? ")
previous_label.grid(row=16, column = 0, sticky = tk.W)

# poutcome
poutcome_label = ttk.Label(win,text = "What was the outcome of previous campaign for this client? ")
poutcome_label.grid(row=17, column = 0, sticky = tk.W)

# Creating entry box for every label
# Age entry box
age_var = tk.StringVar()
age_entrybox = ttk.Entry(win,width = 5, textvariable= age_var)
age_entrybox.grid(row=0, column =1)

# Job entry box
job_var = tk.StringVar()
job_combobox = ttk.Combobox(win,width = 13, textvariable = job_var, state="readonly")
job_combobox['values'] = ('unemployed', 'services', 'management', 'blue-collar', 
                          'self-employed', 'technician', 'entrepreneur','admin.',
                          'student', 'housemaid', 'retired','unknown')
job_combobox.current(0)
job_combobox.grid(row = 1, column=1)

# Marital entry box
mar_var = tk.StringVar()
mar_combobox = ttk.Combobox(win, width = 13, textvariable = mar_var, state = "readonly")
mar_combobox['values'] = ('married', 'single', 'divorced')
mar_combobox.current(0)
mar_combobox.grid(row = 2, column = 1)

# Education
edu_var = tk.StringVar()
edu_combobox = ttk.Combobox(win, width = 13, textvariable = edu_var, state = "readonly")
edu_combobox['values']= ('primary', 'secondary', 'tertiary', 'unknown')
edu_combobox.current(0)
edu_combobox.grid(row = 3, column = 1)

# Default Status
def_var = tk.StringVar()
def_radio_1 = ttk.Radiobutton(win, text = "Yes", variable = def_var, value = "Yes").grid(row = 5, column = 0)
def_radio_2 = ttk.Radiobutton(win, text = "No", variable = def_var, value = "No").grid(row = 5, column = 1)

# Balance entry box
bal_var = tk.StringVar()
bal_entrybox = ttk.Entry(win,width = 5, textvariable= bal_var)
bal_entrybox.grid(row=6, column =1)

# Housing Status
house_var = tk.StringVar()
house_radio_1 = ttk.Radiobutton(win, text = "Yes", variable = house_var, value = "Yes").grid(row = 8, column = 0)
house_radio_2 = ttk.Radiobutton(win, text = "No", variable = house_var, value = "No").grid(row = 8, column = 1)

# Loan Status
loan_var = tk.StringVar()
loan_radio_1 = ttk.Radiobutton(win, text = "Yes", variable = loan_var, value = "Yes").grid(row = 10, column = 0)
loan_radio_2 = ttk.Radiobutton(win, text = "No", variable = loan_var, value = "No").grid(row = 10, column = 1)

# Contact Method
contact_var = tk.StringVar()
contact_combobox = ttk.Combobox(win, width = 10, textvariable = contact_var, state = "readonly")
contact_combobox['values']= ('cellular', 'unknown', 'telephone')
contact_combobox.current(0)
contact_combobox.grid(row = 11, column = 1)


# Date picker via Calendar
date_var = tk.StringVar()
date_cal = DateEntry(win,width = 10,textvariable = date_var)
date_cal.grid(row = 12, column = 1)

# Duration entry box
duration_var = tk.StringVar()
duration_entrybox = ttk.Entry(win,width = 5, textvariable= duration_var)
duration_entrybox.grid(row=13, column =1)

# Campaign entry box
campaign_var = tk.StringVar()
campaign_entrybox = ttk.Entry(win,width = 5, textvariable= campaign_var)
campaign_entrybox.grid(row=14, column =1)

# pdays entry box
pdays_var = tk.StringVar()
pdays_entrybox = ttk.Entry(win,width = 5, textvariable= pdays_var)
pdays_entrybox.grid(row=15, column =1)

# Previous entry box
previous_var = tk.StringVar()
previous_entrybox = ttk.Entry(win,width = 5, textvariable= previous_var)
previous_entrybox.grid(row=16, column =1)

# Poutcome
poutcome_var = tk.StringVar()
poutcome_combobox = ttk.Combobox(win, width = 10, textvariable = poutcome_var, state = "readonly")
poutcome_combobox['values']= ('unknown', 'failure', 'other', 'success')
poutcome_combobox.current(0)
poutcome_combobox.grid(row = 17, column = 1)
```

```python
#Create button code action function

def action():
    userage = age_var.get()
    userjob = job_var.get()
    usermar = mar_var.get()
    useredu = edu_var.get()
    userdef = def_var.get()
    userbal = bal_var.get()
    userhouse = house_var.get()
    userloan = loan_var.get()
    usercontact = contact_var.get()
    userdate = date_var.get()
    
    userday = userdate.split('/')[1]
    usermonth = userdate.split('/')[0]
    
    userduration = duration_var.get()
    
    usercampaign = campaign_var.get()
    userpdays = pdays_var.get()
    userprevious = previous_var.get()
    userpoutcome = poutcome_var.get()
    

    #write to csv file code here
    with open('user_data.csv', 'a', newline = '') as f:
        dict_writer = DictWriter(f, fieldnames=['age', 'job','marital'
                                                ,'education','default','balance'
                                                ,'housing','loan','contact','day'
                                                ,'month','duration','campaign','pdays'
                                                ,'previous','poutcome'])
        if os.stat('user_data.csv').st_size == 0:        #if file is not empty than header write else not
            dict_writer.writeheader()
       
        dict_writer.writerow({
            'age' : userage,
            'job' : userjob,
            'marital' : usermar,
            'education':useredu,
            'default':userdef,
            'balance': userbal,
            'housing': userhouse,
            'loan': userloan,
            'contact':usercontact,
            'day': userday,
            'month':usermonth,
            'duration':userduration,
            'campaign':usercampaign,
            'pdays': userpdays,
            'previous':userprevious,
            'poutcome':userpoutcome
        })

#     Change color after submit button

    age_label.configure(foreground = 'Blue')
    job_label.configure(foreground = 'Blue')
    mar_label.configure(foreground = 'Blue')
    edu_label.configure(foreground = 'Blue')
    def_label.configure(foreground = 'Blue')
    bal_label.configure(foreground = 'Blue')
    house_label.configure(foreground = 'Blue')
    loan_label.configure(foreground = 'Blue')
    contact_label.configure(foreground = 'Blue')
    date_label.configure(foreground = 'Blue')
    duration_label.configure(foreground = 'Blue')
    campaign_label.configure(foreground = 'Blue')
    pdays_label.configure(foreground = 'Blue')
    previous_label.configure(foreground = 'Blue')
    poutcome_label.configure(foreground = 'Blue')
```

```python
#submit button
submit_button = ttk.Button(win, text = "Submit", command = action)  
submit_button.grid(row=18, column=1)
win.mainloop()      
```

```python
win.quit()
```
